```python
import pandas as pd

import matplotlib.pyplot as plt

# Carregar o dataset paises.csv
df = pd.read_csv('paises.csv', delimiter=';')

# Filtrar os países da América do Norte
north_america = df[df['Region'] == 'North America']

# Traçar os gráficos de linhas
plt.figure(figsize=(10, 6))
plt.plot(north_america['Country'], north_america['Deathrate'], label='Taxa de Mortalidade', marker='o')
plt.plot(north_america['Country'], north_america['Birthrate'], label='Taxa de Natalidade', marker='o')

# Configurar o gráfico
plt.title('Taxa de Mortalidade e Natalidade - América do Norte')
plt.xlabel('Países')
plt.ylabel('Taxas')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Mostrar o gráfico
plt.show()



```

### 2-


```python
space_df = pd.read_csv('space.csv', delimiter=';')

# Filtrar as empresas dos EUA e da China
usa_china = space_df[space_df['Location'].str.contains('USA|China')]

# Extrair o nome do país da coluna 'Location'
usa_china['Country'] = usa_china['Location'].apply(lambda x: 'USA' if 'USA' in x else 'China')

# Remover resultados repetidos
unique_companies = usa_china.drop_duplicates(subset=['Company Name', 'Country'])

# Contar o número de empresas por país
company_counts = unique_companies['Country'].value_counts()

# Plotar o gráfico de barras
company_counts.plot(kind='bar')
plt.title('Número de Empresas Espaciais nos EUA e na China')
plt.xlabel('País')
plt.ylabel('Número de Empresas')
plt.show()
```

### 3-

```python
# Filtrar os dados da empresa Roscosmos
roscosmos_df = space_df[space_df['Company Name'] == 'Roscosmos']

# Contar o número de missões bem-sucedidas e malsucedidas
mission_counts = roscosmos_df['Status Mission'].value_counts()

# Criar o gráfico de torta
plt.figure(figsize=(8, 6))
plt.pie(mission_counts, labels=mission_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Porcentagem de Missões da Roscosmos por Status')
plt.ylabel('Status da Missão')
plt.show()
```
